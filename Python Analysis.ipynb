### Exploratory Data Analysis (EDA) of cc_data

```python
#Importing libraries
import pandas as pd
import numpy as np
import datetime as dt
from datetime import datetime, timedelta
from datetime import date
import statistics as stat
from pandasql import sqldf
import sqlite3
from scipy import stats as scstats
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly as plot 
```

```python
#Importing data
credit = pd.read_csv('cc_data (1).csv')
```

```python
#Data quality checks using custom library
import dataquality as dq

dq.data_check(credit)
```

```python
#Create copy of dataframe
credit_copy = credit
```

```python
#Converting trans_date_trans_time data type to datetime object
credit['trans_date_trans_time'] = pd.to_datetime(credit['trans_date_trans_time'])
credit.head()
```

```python
#Converting dob data type to datetime object
credit['dob'] = pd.to_datetime(credit['dob'])
credit.head()

#Some credit card numbers have additional 0's added to the end of the number.
```

```python
#Rechecking datatypes after data type conversions
credit.info()
```

```python
#Adding separate columns for transaction month, year, and year-month
credit['trans_month'] = credit['trans_date_trans_time'].dt.strftime('%b')
credit['trans_year'] = credit['trans_date_trans_time'].dt.strftime('%Y')
credit['trans_year_month'] = credit['trans_date_trans_time'].dt.strftime('%Y-%b')
credit.head()
```

```python
#Reconverting trans_date_trans_time data type to datetime object (reversed back to object after I created the different columns for
#transaction month, year, and year-month)
credit['trans_date_trans_time'] = pd.to_datetime(credit['trans_date_trans_time'])

#Reconverting dob data type to datetime object
credit['dob'] = pd.to_datetime(credit['dob'])
credit.head()
```

```python
#Rechecking datatypes
credit.info()
```

```python
#Adding column to categorize whether or not the transaction is fraudulent
credit = sqldf('''SELECT *,
                        CASE
                            WHEN is_fraud IS 0 THEN 'False'
                            WHEN is_fraud IS 1 THEN 'True'
                            END AS 'Fraud'
                        FROM credit''')
credit.head()
```

```python
#Sorting data by transaction date/time and pulling back first 5 rows (looking for first transaction date)
credit_sorted = credit.sort_values(by = ['trans_date_trans_time'])
credit_sorted.head()
```

```python
#Looking for last transaction date
credit_sorted.tail()
```

```python
#Distribution of transaction amount
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'amt')

plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')

plt.show()
```

```python
#Distribution of latitude of the transaction
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'lat')

plt.xlabel('Transaction Latitude')
plt.ylabel('Frequency')

plt.show()
```

```python
#Distribution of longitude of the transaction
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'long')

plt.xlabel('Transaction Longitude')
plt.ylabel('Frequency')

plt.show()
```

```python
#Distribution of city population where transaction occurred
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'city_pop')

plt.xlabel('City Population')
plt.ylabel('Frequency')

plt.show()
```

```python
#Distribution of unix time
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'unix_time')

plt.xlabel('Unix Time')
plt.ylabel('Frequency')

plt.show()
```

```python
#Distribution of merchant latitude
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'merch_lat')

plt.xlabel('Merchant Latitude')
plt.ylabel('Frequency')

plt.show()
```

```python
#Distribution of merchant longitude
plt.figure(figsize = (15, 5))

sns.histplot(data = credit, x = 'merch_long')

plt.xlabel('Merchant Longitude')
plt.ylabel('Frequency')

plt.show()
```

```python
#looking for correlation between transaction amount and city population
sns.scatterplot(data = credit, x = 'amt', y = 'city_pop')

plt.xlabel('Transaction Amount')
plt.ylabel('City Population')

plt.show() 

#Calculating correlation coefficient
corr_coeff = credit['amt'].corr(credit['city_pop'])
print('Correlation Coefficient:')
print(corr_coeff) #no correlation
```

```python
#Looking for correlation between transaction amount and whether it's fraud
sns.scatterplot(data = credit, x = 'amt', y = 'is_fraud')

plt.xlabel('Transaction Amount')
plt.ylabel('Is Fraud')

plt.show() 

#Calculating correlation coefficient
corr_coeff = credit['amt'].corr(credit['is_fraud'])
print('Correlation Coefficient:')
print(corr_coeff) #no correlation
```

```python
#Looking at distribution of transaction amounts by whether or not they were considered fraud
plt.figure(figsize = (10, 9))

sns.histplot(data = credit, x = 'amt', hue = 'is_fraud', binwidth = 100, bins = 5)

plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')

#Setting limit on x-axis to zoom in on where the values of the histogram are
plt.xlim(0, 2000)

plt.show() #majority of transactions not considered fraud
```

```python
#Looking for outliers in city population using IQR method
q1 = credit['city_pop'].quantile(0.25)
q3 = credit['city_pop'].quantile(0.75)
IQR = q3 - q1
print('Interquartile Range: ', IQR)

low = q1 - (1.5 * IQR)
print('Lower Fence: ', low)

upper = q3 + (1.5 * IQR)
print('Upper Fence: ', upper)
```

```python
#Looking for number of entries with city population below the lower fence
below_lowfence = credit.query('city_pop < -28634.5')
print('Number of data points below lower fence: ', len(below_lowfence))
```

```python
#Looking for number of entries with city population above the upper fence
above_upperfence = credit.query('city_pop > 49705.5')
print('Number of data points above upper fence: ', len(above_upperfence))
```

```python
#Looking for outliers in transaction amount using IQR method
q1 = credit['amt'].quantile(0.25)
q3 = credit['amt'].quantile(0.75)
IQR = q3 - q1
print('Interquartile Range: ', IQR)

low = q1 - (1.5 * IQR)
print('Lower Fence: ', low)

upper = q3 + (1.5 * IQR)
print('Upper Fence: ', upper)
```

```python
#Looking for number of entries with transaction amount below the lower fence
below_lowfence = credit.query('amt < -100.46625')
print('Number of data points below lower fence: ', len(below_lowfence))
```

```python
#Looking for number of entries with transaction amount above the upper fence
above_upperfence = credit.query('amt > 193.20375')
print('Number of data points above upper fence: ', len(above_upperfence))
```

```python
#Creating line graph to look for trends in transaction amounts over time
plt.figure(figsize = (15, 5))

sns.lineplot(data = credit, x = 'trans_year_month', y = 'amt')

plt.xlabel('Transaction Date')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Transaction Amount')

plt.show()
```

```python
#Looking at distribution of transaction amounts among different categories
plt.figure(figsize = (15, 5))

sns.barplot(data = credit, x = 'category', y = 'amt', hue = 'Fraud')

plt.xlabel('Category')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Transaction Amount')

plt.show()
```

```python
#Creating subset of data for fraudulent transactions
fraud_true = credit.query('is_fraud == 1')
display(fraud_true.head())
print( )
print('Number of Fraudulent Transactions: ', len(fraud_true))
```

```python
#Creating subset of data for non-fraudulent transactions
fraud_false = credit.query('is_fraud == 0')
display(fraud_false.head())
print( )
print('Number of Non-Fraudulent Transactions: ', len(fraud_false))
```

```python
#Looking at distribution of transaction amounts among fraudulent transaction
plt.figure(figsize = (10, 5))

sns.histplot(data = fraud_true, x = 'amt')

plt.title('Fraudulent Transactions')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')

plt.show()


#Looking at distribution of transaction amounts among non-fraudulent transaction
plt.figure(figsize = (10, 5))

sns.histplot(data = fraud_false, x = 'amt', bins = 300, binwidth = 20)

plt.title('Non-Fraudulent Transactions')
plt.xlabel('Transaction Amount')
plt.xlim(0, 1200)
plt.ylabel('Frequency')

plt.show()
```

```python
#Looking at distribution of fraudulent transactions by state
plt.figure(figsize = (15, 5))

sns.histplot(data = fraud_true, x = 'state')

plt.title('Fraudulent Transactions')
plt.xlabel('State')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')

plt.show()


#Looking at distribution of non-fraudulent transactions by state
plt.figure(figsize = (15, 5))

sns.histplot(data = fraud_false, x = 'state')

plt.title('Non-Fraudulent Transactions')
plt.xlabel('State')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')

plt.show()
```

```python
#Looking at distribution of category among fraudulent transaction
plt.figure(figsize = (10, 5))

sns.histplot(data = fraud_true, x = 'category')

plt.title('Fraudulent Transactions')
plt.xlabel('Category')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')

plt.show()


#Looking at distribution of category among non-fraudulent transaction
plt.figure(figsize = (10, 5))

sns.histplot(data = fraud_false, x = 'category')

plt.title('Non-Fraudulent Transactions')
plt.xlabel('Category')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')

plt.show()
```

```python
#Looking at distribution of fraudulent transactions by gender
plt.figure(figsize = (8, 5))

sns.histplot(data = fraud_true, x = 'gender')

plt.title('Fraudulent Transactions')
plt.xlabel('Gender')
plt.ylabel('Frequency')

plt.show()


#Looking at distribution of non-fraudulent transactions by gender
plt.figure(figsize = (8, 5))

sns.histplot(data = fraud_false, x = 'gender')

plt.title('Non-Fraudulent Transactions')
plt.xlabel('Gender')
plt.ylabel('Frequency')

plt.show()
```

```python
#Creating line graph to look for trends in fraudulent transaction amounts over time
plt.figure(figsize = (15, 5))

sns.lineplot(data = fraud_true, x = 'trans_year_month', y = 'amt')

plt.title('Fraudulent Transaction Amounts Over Time')
plt.xlabel('Transaction Date')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Transaction Amount')

plt.show()



#Creating line graph to look for trends in non-fraudulent transaction amounts over time
plt.figure(figsize = (15, 5))

sns.lineplot(data = fraud_false, x = 'trans_year_month', y = 'amt')

plt.title('Non-Fraudulent Transaction Amounts Over Time')
plt.xlabel('Transaction Date')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Transaction Amount')

plt.show()
```
